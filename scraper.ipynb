```python
from selenium import webdriver
import time
import pandas as pd
import numpy as np
import os
import csv
```

```python
def get_data():
    url = 'https://www.argentina.gob.ar/precios-cuidados/listado-de-productos'
    drive = webdriver.Chrome() 
    drive.get(url)

    td = [1,2,5]
    item_name = []
    item_category = []
    
    item_price_AMBA = []
    item_price_restoPBA = []
    item_price_CCL = []
    item_price_neaNOA = []
    item_price_Patagonia = []
    
    aux = True
    count = 0
    count_region = 1
    
    
    while aux==True:
        long = len(drive.find_elements_by_xpath("//*[@id='ponchoTable']/tbody/tr"))
        for i in range(1,long+1):
            if count_region == 1:
                name = drive.find_element_by_xpath("//*[@id='ponchoTable']/tbody/tr[{}]/td[{}]/p".format(i,td[0])).text
                item_name.append(name)
                category = drive.find_element_by_xpath("//*[@id='ponchoTable']/tbody/tr[{}]/td[{}]/p".format(i,td[1])).text
                item_category.append(category)
                try:
                    price = drive.find_element_by_xpath("//*[@id='ponchoTable']/tbody/tr[{}]/td[{}]/p".format(i,td[2])).text
                    item_price_AMBA.append(price)
                except:
                    item_price_AMBA.append(0)
            elif count_region==2:
                try:
                    price = drive.find_element_by_xpath("//*[@id='ponchoTable']/tbody/tr[{}]/td[{}]/p".format(i,td[2])).text
                    item_price_restoPBA.append(price)
                except:
                    item_price_restoPBA.append(0)
            elif count_region==3:
                try:
                    price = drive.find_element_by_xpath("//*[@id='ponchoTable']/tbody/tr[{}]/td[{}]/p".format(i,td[2])).text
                    item_price_CCL.append(price)
                except:
                    item_price_CCL.append(0)
            elif count_region==4:
                try:
                    price = drive.find_element_by_xpath("//*[@id='ponchoTable']/tbody/tr[{}]/td[{}]/p".format(i,td[2])).text
                    item_price_neaNOA.append(price)
                except:
                    item_price_neaNOA.append(0)
            elif count_region==5:
                try:
                    price = drive.find_element_by_xpath("//*[@id='ponchoTable']/tbody/tr[{}]/td[{}]/p".format(i,td[2])).text
                    item_price_Patagonia.append(price)
                except:
                    item_price_Patagonia.append(0)

        nexta = drive.find_element_by_xpath("//*[@id='ponchoTable_next']/a").click()

        count += 1

        if count == 31:
            print('Terminó la vuelta número {}'.format(count_region))
            count = 0

            count_region += 1
            if count_region != 6:
                next_region = drive.find_element_by_xpath("//*[@id='selectZona']/option[{}]".format(count_region)).click()
                first_page = drive.find_element_by_xpath("//*[@id='ponchoTable_paginate']/ul/li[2]/a").click()
            else:
                aux = False

    drive.close()
    
    data_dic = {'item_name':item_name,
                'item_category':item_category,
                'item_price':[item_price_AMBA, item_price_restoPBA, item_price_CCL, item_price_neaNOA, item_price_Patagonia]
               }
    
    return data_dic
```

```python
def save_data(data_dic):
    df = pd.DataFrame(data=
                      {'item_name':data_dic['item_name'],
                       'item_category':data_dic['item_category'],
                       'item_price_AMBA':data_dic['item_price'][0],
                       'item_price_restoPBA':data_dic['item_price'][1],
                       'item_price_CCL':data_dic['item_price'][2],
                       'item_price_neaNOA':data_dic['item_price'][3],
                       'item_price_Patagonia':data_dic['item_price'][4],
                      })
    df.to_csv('data/listadeprecios.csv',sep=',',index=False)
    return df
```

```python
data_dic = get_data()
df = save_data(data_dic)
```
